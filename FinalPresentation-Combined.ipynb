```python
import pandas as pd

# Assuming DataStandardizer class is already defined as mentioned in previous messages

# Load your data
employee_performance = pd.read_csv("./data/Extended_Employee_Performance_and_Productivity_Data.csv")
salary_data = pd.read_csv("./data/Salary_Data.csv")
recruitment_data = pd.read_csv("./data/recruitment_dataset.csv")

# Check column names of the dataframes
print("Employee Performance Columns:", employee_performance.columns)
print("Salary Data Columns:", salary_data.columns)
print("Recruitment Data Columns:", recruitment_data.columns)

class DataStandardizer:
    def __init__(self, employee_performance, salary_data, recruitment_data):
        self.employee_performance = employee_performance
        self.salary_data = salary_data
        self.recruitment_data = recruitment_data
    
    def standardize_column_names(self, df):
        """Standardize column names across datasets by making them lower case and removing spaces."""
        df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
        return df
    
    def standardize_education_level(self, df):
        """Standardize education level values across all datasets."""
        # Check if 'education_level' exists in the dataframe before applying the transformation
        if 'education_level' in df.columns:
            education_mapping = {
                'high school': 'High School',
                "bachelor's degree": 'Bachelor\'s Degree',
                'master': "Master's Degree",
                'phd': 'PhD',
                'doctorate': 'PhD'
            }
            
            # Apply mapping and ensure uniform casing
            df['education_level'] = df['education_level'].str.lower().map(education_mapping).fillna(df['education_level'])
            df['education_level'] = df['education_level'].str.title()  # Standardize the casing
        return df
    
    def standardize(self):
        """Apply the standardization steps across all datasets."""
        # Standardize column names
        self.employee_performance = self.standardize_column_names(self.employee_performance)
        self.salary_data = self.standardize_column_names(self.salary_data)
        self.recruitment_data = self.standardize_column_names(self.recruitment_data)
        
        # Standardize education level
        self.employee_performance = self.standardize_education_level(self.employee_performance)
        self.salary_data = self.standardize_education_level(self.salary_data)
        self.recruitment_data = self.standardize_education_level(self.recruitment_data)
        
        return self.employee_performance, self.salary_data, self.recruitment_data

```

```python
import pandas as pd

class DataCleaner:
    def __init__(self, employee_performance, salary_data, recruitment_data):
        self.employee_performance = employee_performance
        self.salary_data = salary_data
        self.recruitment_data = recruitment_data
    
    def handle_missing_values(self, df):
        """Fill missing values in each dataset based on context."""
        # For numerical columns, we can fill missing values with the median
        for col in df.select_dtypes(include='number').columns:
            df[col] = df[col].fillna(df[col].median())
        
        # For categorical columns, fill missing values with the mode
        for col in df.select_dtypes(include='object').columns:
            df[col] = df[col].fillna(df[col].mode()[0])
        
        return df
    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats

def plot_histograms_qq_plots(df, columns, group_by):
    """
    Function to plot histograms and Q-Q plots for selected columns and group by a specific feature.

    Parameters:
    - df: DataFrame containing the data.
    - columns: List of column names to plot.
    - group_by: Column name by which to group the data (e.g., 'gender', 'department').
    """
    # Group data by the specified group_by column
    unique_groups = df[group_by].unique()
    
    # Set up the plot
    fig, axes = plt.subplots(len(columns), len(unique_groups)*2, figsize=(24, 18))
    
    for i, col in enumerate(columns):
        for j, group in enumerate(unique_groups):
            # Group data for each unique value in group_by
            group_data = df[df[group_by] == group]
            
            # Histogram for the group
            sns.histplot(group_data[col], kde=True, ax=axes[i, j*2], label=group)
            axes[i, j*2].set_title(f'Histogram of {col} ({group})')
            
            # Q-Q plot for the group
            stats.probplot(group_data[col].dropna(), dist="norm", plot=axes[i, j*2+1])
            axes[i, j*2+1].set_title(f'Q-Q plot of {col} ({group})')
            
            # Add legends
            axes[i, j*2].legend()

    plt.tight_layout()
    plt.show()
```

### 2. T-Tests and F-Tests


```python
from scipy.stats import ttest_ind

def perform_t_test(df, column, group_by):
    """
    Perform a T-test between two groups in a dataframe.

    Parameters:
    - df: DataFrame containing the data.
    - column: The column on which to perform the T-test.
    - group_by: The column by which to group the data (e.g., 'gender').

    Returns:
    - t_stat: T-statistic value.
    - p_value: P-value of the test.
    """
    group1 = df[df[group_by] == df[group_by].unique()[0]]
    group2 = df[df[group_by] == df[group_by].unique()[1]]

    t_stat, p_value = ttest_ind(group1[column], group2[column])
    return t_stat, p_value

# Example usage
t_stat, p_value = perform_t_test(employee_performance, 'performance_score', 'gender')
print(f"T-statistic: {t_stat}, P-value: {p_value}")
t_stat, p_value = perform_t_test(employee_performance, 'performance_score', 'resigned')
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

The t-tests show no significant difference in performance scores based on gender (p = 0.30) or resignation status (p = 0.39).

```python
from scipy.stats import f_oneway


def perform_f_test(df, column, group_by):
    """
    Perform an F-test (ANOVA) to compare the means of a specified column across different groups.

    Parameters:
    - df: DataFrame containing the data.
    - column: The column on which to perform the F-test.
    - group_by: The column by which to group the data (e.g., 'education_level').

    Returns:
    - f_stat: F-statistic value.
    - p_value: P-value of the test.
    """
    groups = df[group_by].unique()
    grouped_data = [df[df[group_by] == group][column] for group in groups]
    f_stat, p_value = f_oneway(*grouped_data)
    return f_stat, p_value


f_stat, p_value = perform_f_test(employee_performance, 'monthly_salary', 'gender')
print(f"F-statistic: {f_stat}, P-value: {p_value}")
f_stat, p_value = perform_f_test(employee_performance, 'monthly_salary', 'education_level')
print(f"F-statistic: {f_stat}, P-value: {p_value}")
```

The F-statistic of 0.28 and p-value of 0.76 for the comparison between monthly salary and gender suggest that there is no significant difference in the salary variances between genders. Similarly, the F-statistic of 1.53 and p-value of 0.20 for the comparison between monthly salary and education level indicate no significant difference in salary variances across different education levels.

### 3. Chi-Square Test: Gender and Decision

The objective of this analysis is to determine whether there is a significant association between **gender** and **decision**. Both variables are categorical, so we can use the **Chi-Square Test of Independence** to assess the relationship.

## Chi-Square Test of Independence:
The Chi-Square Test of Independence is used to determine whether two categorical variables are independent or if they are related. The test compares the observed frequency distribution of the data with the expected frequency distribution under the assumption that the variables are independent.

### Null Hypothesis (H₀):
- **H₀**: There is no association between **gender** and **decision**. That is, gender and decision are independent variables.
  
### Alternative Hypothesis (H₁):
- **H₁**: There is an association between **gender** and **decision**. That is, gender and decision are not independent.


```python
import pandas as pd
from scipy.stats import chi2_contingency
from IPython.display import display

def chi_square_tests_by_company(df):
    # Split the dataframe into subsets for each company
    company_A_data = df[df['company'] == 'A']
    company_B_data = df[df['company'] == 'B']
    company_C_data = df[df['company'] == 'C']
    company_D_data = df[df['company'] == 'D']

    # Function to perform Chi-Square Test and return results
    def chi_square_test(data):
        contingency_table = pd.crosstab(data['gender'], data['decision'])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        return chi2, p, dof, expected

    # Perform Chi-Square tests for each company
    results = []

    for company, data in zip(['Entire Dataset','A', 'B', 'C', 'D'], [recruitment_data, company_A_data, company_B_data, company_C_data, company_D_data]):
        chi2, p, dof, expected = chi_square_test(data)
        results.append({
            'Company': company,
            'Chi-Square Statistic': chi2,
            'P-value': p,
            'Degrees of Freedom': dof,
            'Expected Frequencies': expected
        })

    # Convert the results into a DataFrame for a clear table format
    results_df = pd.DataFrame(results)

    # Return the results DataFrame
    return results_df

result_df = chi_square_tests_by_company(recruitment_data)

# Pretty print the results in a table format
display(result_df)
```

> In summary, the Chi-Square test indicates gender bias in the overall dataset, with a significant relationship between gender and the decision variable.

> Company B and Company D show significant gender bias in recruitment decisions, as their p-values are less than 0.05.

> Company A and Company C do not show any significant association between gender and decision, indicating no gender bias in their recruitment processes.

### 4. ANOVA


```python
import pandas as pd
from scipy.stats import f_oneway

def perform_anova_on_hiring_decision(data):
    """
    Perform ANOVA to see if gender has an impact on hiring decisions for the entire dataset and each company.

    Parameters:
    - data: DataFrame containing the recruitment data.

    Returns:
    - results_df: DataFrame containing the ANOVA results for the entire dataset and each company.
    """
    # Split the dataset based on companies
    company_a = data[data['company'] == 'A']
    company_b = data[data['company'] == 'B']
    company_c = data[data['company'] == 'C']
    company_d = data[data['company'] == 'D']

    # Group the data by gender for the entire dataset
    groups = [data[data['gender'] == gender]['decision'] for gender in data['gender'].unique()]

    # Perform ANOVA for the entire dataset
    f_stat, p_value = f_oneway(*groups)
    results = [{'Company': 'Entire Dataset', 'F-statistic': f_stat, 'P-value': p_value}]

    # Perform ANOVA for each company
    for company, df in zip(['A', 'B', 'C', 'D'], [company_a, company_b, company_c, company_d]):
        groups = [df[df['gender'] == gender]['decision'] for gender in df['gender'].unique()]
        f_stat, p_value = f_oneway(*groups)
        results.append({'Company': company, 'F-statistic': f_stat, 'P-value': p_value})

    # Convert results to DataFrame
    results_df = pd.DataFrame(results)

    return results_df

# Example usage
results_df = perform_anova_on_hiring_decision(recruitment_data)
display(results_df)
```

Companies B and D exhibit significant bias in hiring decisions based on gender, while Companies A and C do not show significant differences.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Ensure column names are standardized
employee_performance = standardizer.standardize_column_names(employee_performance)

# Check if 'decision' column exists
if 'decision' not in employee_performance.columns:
    raise KeyError("'decision' column is missing from the dataset")

def logistic_regression_analysis(employee_performance):
    # Convert decision into binary (e.g., 1 for hired, 0 for not hired)
    employee_performance['decision'] = employee_performance['decision'].map({True: 1, False: 0})

    # Select features and target
    X = employee_performance[['age', 'gender', 'education_level']]
    y = employee_performance['decision']

    # Split data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Apply Logistic Regression
    classifier = LogisticRegression()
    classifier.fit(X_train, y_train)

    # Predict the hiring decision
    predictions = classifier.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, predictions)
    report = classification_report(y_test, predictions)

    return accuracy, report

def logistic_regression_by_gender(employee_performance):
    # Split the data into male and female
    male_data = employee_performance[employee_performance['gender'] == 'Male']
    female_data = employee_performance[employee_performance['gender'] == 'Female']

    # Perform logistic regression analysis for male employees
    male_accuracy, male_report = logistic_regression_analysis(male_data)

    # Perform logistic regression analysis for female employees
    female_accuracy, female_report = logistic_regression_analysis(female_data)

    return (male_accuracy, male_report), (female_accuracy, female_report)

# Example usage
overall_accuracy, overall_report = logistic_regression_analysis(employee_performance)
(male_accuracy, male_report), (female_accuracy, female_report) = logistic_regression_by_gender(employee_performance)

print("Overall Accuracy:", overall_accuracy)
print("Overall Classification Report:\n", overall_report)
print("Male Accuracy:", male_accuracy)
print("Male Classification Report:\n", male_report)
print("Female Accuracy:", female_accuracy)
print("Female Classification Report:\n", female_report)
```

5. Probabilistic Reasoning (e.g., Naive Bayes)

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder


def naive_bayes_analysis(employee_performance):

    # Encode categorical variables
    le = LabelEncoder()
    employee_performance['gender'] = le.fit_transform(employee_performance['gender'])
    employee_performance['education_level'] = le.fit_transform(employee_performance['education_level'])

    # Select features and target
    X = employee_performance[['age', 'gender', 'education_level']]
    y = employee_performance['decision']

    # Apply Naive Bayes Classification
    nb_classifier = GaussianNB()
    nb_classifier.fit(X, y)

    # Predict the hiring decision probabilities
    probabilities = nb_classifier.predict_proba(X)
    return probabilities

def naive_bayes_by_gender(employee_performance):
    # Split the data into male and female
    male_data = employee_performance[employee_performance['gender'] == 'Male']
    female_data = employee_performance[employee_performance['gender'] == 'Female']

    # Perform Naive Bayes analysis for male employees
    male_probabilities = naive_bayes_analysis(male_data)

    # Perform Naive Bayes analysis for female employees
    female_probabilities = naive_bayes_analysis(female_data)

    return male_probabilities, female_probabilities

# Example usage
probabilities = naive_bayes_analysis(employee_performance)
male_probabilities, female_probabilities = naive_bayes_by_gender(employee_performance)
```
