## Step 2 - Code Implementation

```python
import pandas as pd

# Assuming DataStandardizer class is already defined as mentioned in previous messages

# Load your data
employee_performance = pd.read_csv("./data/Extended_Employee_Performance_and_Productivity_Data.csv")
salary_data = pd.read_csv("./data/Salary_Data.csv")
recruitment_data = pd.read_csv("./data/recruitment_dataset.csv")

# Check column names of the dataframes
print("Employee Performance Columns:", employee_performance.columns)
print("Salary Data Columns:", salary_data.columns)
print("Recruitment Data Columns:", recruitment_data.columns)

class DataStandardizer:
    def __init__(self, employee_performance, salary_data, recruitment_data):
        self.employee_performance = employee_performance
        self.salary_data = salary_data
        self.recruitment_data = recruitment_data
    
    def standardize_column_names(self, df):
        """Standardize column names across datasets by making them lower case and removing spaces."""
        df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
        return df
    
    def standardize_education_level(self, df):
        """Standardize education level values across all datasets."""
        # Check if 'education_level' exists in the dataframe before applying the transformation
        if 'education_level' in df.columns:
            education_mapping = {
                'high school': 'High School',
                "bachelor's degree": 'Bachelor\'s Degree',
                'master': "Master's Degree",
                'phd': 'PhD',
                'doctorate': 'PhD'
            }
            
            # Apply mapping and ensure uniform casing
            df['education_level'] = df['education_level'].str.lower().map(education_mapping).fillna(df['education_level'])
            df['education_level'] = df['education_level'].str.title()  # Standardize the casing
        return df
    
    def standardize(self):
        """Apply the standardization steps across all datasets."""
        # Standardize column names
        self.employee_performance = self.standardize_column_names(self.employee_performance)
        self.salary_data = self.standardize_column_names(self.salary_data)
        self.recruitment_data = self.standardize_column_names(self.recruitment_data)
        
        # Standardize education level
        self.employee_performance = self.standardize_education_level(self.employee_performance)
        self.salary_data = self.standardize_education_level(self.salary_data)
        self.recruitment_data = self.standardize_education_level(self.recruitment_data)
        
        return self.employee_performance, self.salary_data, self.recruitment_data

```

```python
import pandas as pd

class DataCleaner:
    def __init__(self, employee_performance, salary_data, recruitment_data):
        self.employee_performance = employee_performance
        self.salary_data = salary_data
        self.recruitment_data = recruitment_data
    
    def handle_missing_values(self, df):
        """Fill missing values in each dataset based on context."""
        # For numerical columns, we can fill missing values with the median
        for col in df.select_dtypes(include='number').columns:
            df[col] = df[col].fillna(df[col].median())
        
        # For categorical columns, fill missing values with the mode
        for col in df.select_dtypes(include='object').columns:
            df[col] = df[col].fillna(df[col].mode()[0])
        
        return df
    
```

```python
# Initialize the standardizer
standardizer = DataStandardizer(employee_performance, salary_data, recruitment_data)

# Apply the standardization
employee_performance, salary_data, recruitment_data = standardizer.standardize()

# Now, all datasets are standardized with common column names and consistent values
print("Employee Performance Columns After Standardization:", employee_performance.columns)
print("Salary Data Columns After Standardization:", salary_data.columns)
print("Recruitment Data Columns After Standardization:", recruitment_data.columns)

# Optionally, print the first few rows to verify
print(employee_performance.head())
print(salary_data.head())
print(recruitment_data.head())
```

```python
# Initialize the cleaner
cleaner = DataCleaner(employee_performance, salary_data, recruitment_data)

# Apply the cleaning
cleaner.handle_missing_values(employee_performance)
cleaner.handle_missing_values(salary_data)
cleaner.handle_missing_values(recruitment_data)

# Optionally, print the first few rows to verify the changes
print(employee_performance.head())
print(salary_data.head())
print(recruitment_data.head())
```

### 1. Normality Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats

def plot_histograms_qq_plots(df, columns, group_by):
    """
    Function to plot histograms and Q-Q plots for selected columns and group by a specific feature.

    Parameters:
    - df: DataFrame containing the data.
    - columns: List of column names to plot.
    - group_by: Column name by which to group the data (e.g., 'gender', 'department').
    """
    # Group data by the specified group_by column
    unique_groups = df[group_by].unique()
    
    # Set up the plot
    fig, axes = plt.subplots(len(columns), len(unique_groups)*2, figsize=(24, 18))
    
    for i, col in enumerate(columns):
        for j, group in enumerate(unique_groups):
            # Group data for each unique value in group_by
            group_data = df[df[group_by] == group]
            
            # Histogram for the group
            sns.histplot(group_data[col], kde=True, ax=axes[i, j*2], label=group)
            axes[i, j*2].set_title(f'Histogram of {col} ({group})')
            
            # Q-Q plot for the group
            stats.probplot(group_data[col].dropna(), dist="norm", plot=axes[i, j*2+1])
            axes[i, j*2+1].set_title(f'Q-Q plot of {col} ({group})')
            
            # Add legends
            axes[i, j*2].legend()

    plt.tight_layout()
    plt.show()
```

### 2. Chi-Square Test: Gender and Decision

The objective of this analysis is to determine whether there is a significant association between **gender** and **decision**. Both variables are categorical, so we can use the **Chi-Square Test of Independence** to assess the relationship.

## Chi-Square Test of Independence:
The Chi-Square Test of Independence is used to determine whether two categorical variables are independent or if they are related. The test compares the observed frequency distribution of the data with the expected frequency distribution under the assumption that the variables are independent.

### Null Hypothesis (H₀):
- **H₀**: There is no association between **gender** and **decision**. That is, gender and decision are independent variables.
  
### Alternative Hypothesis (H₁):
- **H₁**: There is an association between **gender** and **decision**. That is, gender and decision are not independent.


```python
import pandas as pd
from scipy.stats import chi2_contingency
from IPython.display import display

def chi_square_tests_by_company(df):
    # Split the dataframe into subsets for each company
    company_A_data = df[df['company'] == 'A']
    company_B_data = df[df['company'] == 'B']
    company_C_data = df[df['company'] == 'C']
    company_D_data = df[df['company'] == 'D']

    # Function to perform Chi-Square Test and return results
    def chi_square_test(data):
        contingency_table = pd.crosstab(data['gender'], data['decision'])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        return chi2, p, dof, expected

    # Perform Chi-Square tests for each company
    results = []

    for company, data in zip(['Entire Dataset','A', 'B', 'C', 'D'], [recruitment_data, company_A_data, company_B_data, company_C_data, company_D_data]):
        chi2, p, dof, expected = chi_square_test(data)
        results.append({
            'Company': company,
            'Chi-Square Statistic': chi2,
            'P-value': p,
            'Degrees of Freedom': dof,
            'Expected Frequencies': expected
        })

    # Convert the results into a DataFrame for a clear table format
    results_df = pd.DataFrame(results)

    # Return the results DataFrame
    return results_df

result_df = chi_square_tests_by_company(recruitment_data)

# Pretty print the results in a table format
display(result_df)
```

> In summary, the Chi-Square test indicates gender bias in the overall dataset, with a significant relationship between gender and the decision variable.

> Company B and Company D show significant gender bias in recruitment decisions, as their p-values are less than 0.05.

> Company A and Company C do not show any significant association between gender and decision, indicating no gender bias in their recruitment processes.

## Step 4 - Dimensionality reduction

Principal Component Analysis (PCA)

PCA reduces the dimensionality of numerical features by transforming them into a set of orthogonal principal components.

Steps:

- Select numerical columns (e.g., age, monthly_salary, years_at_company).
- Normalize the data using StandardScaler.
- Apply PCA to extract the principal components.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Select numerical columns
numerical_cols = ['age', 'years_at_company', 'monthly_salary', 
                  'work_hours_per_week', 'projects_handled', 
                  'overtime_hours', 'sick_days', 'training_hours']

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(employee_performance[numerical_cols])

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
principal_components = pca.fit_transform(scaled_data)

# Add principal components to the dataset
employee_performance['PCA1'] = principal_components[:, 0]
employee_performance['PCA2'] = principal_components[:, 1]

# Explained variance
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Define features and target
X = employee_performance[numerical_cols]
y = employee_performance['resigned']  # Target variable

# Apply LDA
lda = LDA(n_components=1)  # Reduce to 1 dimension
lda_components = lda.fit_transform(X, y)

# Add LDA components to the dataset
employee_performance['LDA1'] = lda_components
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def process_and_plot_job_time_series(df, job_category_mapping, job_department_mapping):
    """
    Process the dataframe to create a time series visualization of job categories and departments.
    
    Parameters:
    df (pandas.DataFrame): Input dataframe with job titles
    job_category_mapping (dict): Mapping of job titles to job categories
    job_department_mapping (dict): Mapping of job categories to job departments
    
    Returns:
    pandas.DataFrame: Processed dataframe with additional columns
    """
    # Create a copy of the dataframe to avoid modifying the original
    processed_df = df.copy()
    
    # Calculate year hired by subtracting years of experience from 2020
    processed_df['year_hired'] = 2020 - processed_df['years_of_experience']
    
    # Map job titles to job categories
    processed_df['job_category'] = processed_df['job_title'].map(job_category_mapping)
    
    # Map job categories to job departments
    processed_df['job_department'] = processed_df['job_category'].map(job_department_mapping)
    
    # Get the top 3 job departments with the highest number of people
    top_departments = processed_df['job_department'].value_counts().nlargest(3).index
    
    # Filter the dataframe to include only the top 3 job departments
    filtered_df = processed_df[processed_df['job_department'].isin(top_departments)]
    
    # Group the data and prepare for visualization
    def plot_time_series(data, title):
        plt.figure(figsize=(15, 7))
        # Count job categories or departments by year hired and gender
        counts = data.groupby(['year_hired', 'gender']).size().unstack(fill_value=0)
        
        # Plot stacked area chart
        counts.plot(kind='line', alpha=0.7, ax=plt.gca())
        
        plt.title(title)
        plt.xlabel('Year Hired')
        plt.ylabel('Number of Employees')
        plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.show()
    
    # Plot for each of the top 3 departments
    for department in top_departments:
        dept_data = filtered_df[filtered_df['job_department'] == department]
        plot_time_series(dept_data, f'Hirings Over Time - {department}')
    
    # Plot for the whole dataset
    plot_time_series(filtered_df, 'Hirings Over Time - Entire Dataset')
    
    return processed_df

# Example usage (you would replace df with your actual dataframe)
result_df = process_and_plot_job_time_series(salary_data, job_category_mapping, job_department_mapping)
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
 
# Class to handle employee data loading and preprocessing
class IterationHelper:
    def __init__(self, file_path, feature_list):
        self.file_path = file_path
        self.data = None
        self.feature_list = feature_list
 
    def performIteration(self):
        employee_data = EmployeeData(self.file_path,
                             self.feature_list)
        data = employee_data.load_data()
        features = employee_data.preprocess_data()
        employee_data.gender_summary_statistics()
        # Step 2: Perform clustering
        clustering = EmployeeClustering(features)
        clustering.scale_features()
        clustering.find_optimal_k()
        clustered_data = clustering.apply_kmeans()
        # Step 3: Visualize the clusters
        for i,x in enumerate(employee_data.feature_list):
            for j in range(i+1, len(employee_data.feature_list)):
                if(x != employee_data.feature_list[j]):
                    clustering.visualize_clusters(x, employee_data.feature_list[j])
   
class EmployeeData:
    def __init__(self, file_path, feature_list):
        self.file_path = file_path
        self.data = None
        self.feature_list = feature_list
 
 
    def load_data(self):
        self.data = pd.read_csv(self.file_path)
        return self.data
 
    def preprocess_data(self):
        # Convert categorical data to numerical (if needed)
        self.data['Gender'] = self.data['Gender'].map({'Male': 0, 'Female': 1})
        # Select features for clustering
        self.features = self.data[self.feature_list]
        return self.features
 
    def gender_summary_statistics(self):
        # Display summary statistics by gender for key metrics
        summary_stats = self.data.groupby('Gender').agg({
            'Monthly_Salary': ['mean', 'median'],
            'Years_At_Company': ['mean', 'median'],
            'Performance_Score': ['mean', 'median'],
            'Promotions': ['mean', 'median'],
            'Employee_Satisfaction_Score': ['mean', 'median']
        })
        print("Summary Statistics by Gender:\n", summary_stats)
       
        # # Visualizations to check distributions by gender
        # plt.figure(figsize=(10, 6))
        # sns.boxplot(x='Gender', y='Monthly_Salary', data=self.data)
        # plt.title("Monthly Salary Distribution by Gender")
        # plt.show()
 
        # plt.figure(figsize=(10, 6))
        # sns.boxplot(x='Gender', y='Performance_Score', data=self.data)
        # plt.title("Performance Score Distribution by Gender")
        # plt.show()
 
class Plotters:
    def plot_scatter(df, x, y, title, x_label, y_label):
        # Plot a scatter plot to visualize the relationship between two variables
        plt.figure(figsize=(10, 6))
        plt.scatter(df[x], df[y])
        plt.title(title)
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        # Calculate and display summary statistics
        x_mean = df[x].mean()
        y_mean = df[y].mean()
        correlation = df[[x, y]].corr().iloc[0, 1]
        summary_text = (
            f"Summary:\n"
            f"Mean of {x}: {x_mean:.2f}\n"
            f"Mean of {y}: {y_mean:.2f}\n"
            f"Correlation between {x} and {y}: {correlation:.2f}"
        )
        plt.figtext(0.15, -0.1, summary_text, wrap=True, fontsize=12, ha='left')
        plt.show()
       
    def plot_scatter_diagram(df, dependent_column_name):
        fig, axes = plt.subplots(4, int((len(df.columns)/4) + 1),figsize=(15,15))
        for i, col_name in enumerate(df.columns):
            if col_name != dependent_column_name:
                sns.scatterplot(x=col_name, y=dependent_column_name, data=df, ax=axes[i%4, i//4])
        plt.tight_layout()
        plt.show()
 
# Class to handle clustering operations
class EmployeeClustering:
    def __init__(self, data):
        self.data = data
        self.scaled_data = None
        self.model = None
        self.optimal_k = None
 
    def scale_features(self):
        scaler = StandardScaler()
        self.scaled_data = scaler.fit_transform(self.data)
        return self.scaled_data
 
    def find_optimal_k(self, max_k=10):
        sse = []
        for k in range(1, max_k + 1):
            kmeans = KMeans(n_clusters=k, random_state=42)
            kmeans.fit(self.scaled_data)
            sse.append(kmeans.inertia_)
       
        plt.plot(range(1, max_k + 1), sse)
        plt.xlabel("Number of Clusters (K)")
        plt.ylabel("SSE")
        plt.title("Elbow Method to Find Optimal K")
        plt.show()
       
        # After identifying optimal K from the plot, assign it
        self.optimal_k = int(input("Enter the optimal number of clusters (K) from the Elbow plot: "))
 
    def apply_kmeans(self):
        self.model = KMeans(n_clusters=self.optimal_k, random_state=42)
        self.data['Cluster'] = self.model.fit_predict(self.scaled_data)
        return self.data
 
    def visualize_clusters(self, x, y):
        plt.figure(figsize=(4, 4))
        sns.scatterplot(
            x=self.data[x],
            y=self.data[y],
            hue=self.data['Cluster'],
            palette='viridis'
        )
        plt.title(f"Employee Clusters in {x} and {y}")
        plt.show()
 
# Class to handle bias analysis
class BiasAnalysis:
    def __init__(self, data):
        self.data = data
 
    def perform_statistical_tests(self):
        # Test for bias in Monthly Salary by Gender
        male_salary = self.data[self.data['Gender'] == 0]['Monthly_Salary']
        female_salary = self.data[self.data['Gender'] == 1]['Monthly_Salary']
        t_stat, p_val = ttest_ind(male_salary, female_salary)
        print("Monthly Salary T-test between Genders: T-statistic =", t_stat, ", P-value =", p_val)
 
        # Test for bias in Performance Score by Gender
        male_performance = self.data[self.data['Gender'] == 0]['Performance_Score']
        female_performance = self.data[self.data['Gender'] == 1]['Performance_Score']
        t_stat, p_val = ttest_ind(male_performance, female_performance)
        print("Performance Score T-test between Genders: T-statistic =", t_stat, ", P-value =", p_val)
 
 
```
